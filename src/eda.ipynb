```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
def plot_hist(df):
    fig, axs = plt.subplots(3, 1, figsize=(15, 10))
    axs[0].hist(df['contrast_score'], bins=len(df))
    axs[0].set_xlabel('Contrast')

    axs[1].hist(df['sharpness_score'], bins=len(df))
    axs[1].set_xlabel('Blur')

    axs[2].hist(df['noise_score'], bins=len(df))
    axs[2].set_xlabel('Noise')
    fig.show()
```

```python
df = pd.read_csv('../data/data.csv')
```

```python
df_good = df[df['class'] == 'good']
df_low_contrast = df[df['class'] == 'low_contrast']
df_blurry = df[df['class'] == 'blurry']
df_noisy = df[df['class'] == 'noisy']
```

```python
plot_hist(df_good)
```

```python
plot_hist(df_low_contrast)
```

```python
plot_hist(df_blurry)
```

```python
plot_hist(df_noisy)
```
