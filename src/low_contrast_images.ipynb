```python
import os
import numpy as np
import cv2
```

```python
path = 'data/low_contrast/testing/INPUT_IMAGES'
img_sums = {}
for idx, image in enumerate(os.listdir(path)):
    img_path = os.path.join(path, image)
    img = cv2.imread(img_path)
    img_sum = img.mean()
    img_sums[img_path] = img_sum
```

```python
sorted_items = sorted(img_sums.items(), key=lambda item: item[1])
sorted_dict = dict(sorted_items)
```

```python
sorted_dict
```

```python
for idx, item in enumerate(sorted_dict):
    if not (idx < 140 or idx > (len(sorted_dict) - 141)):
        os.remove(item)
```
